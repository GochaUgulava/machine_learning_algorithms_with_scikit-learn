```python
import numpy as np
from sklearn import datasets
```

# Load Data

```python
iris = datasets.load_iris()
iris
```

```python
X = iris['data']
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
labels = np.unique(y)
labels
```

# Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
print(np.bincount(y))
print(np.bincount(y_train))
print(np.bincount(y_test))
```

# Model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion='gini', n_estimators=25, n_jobs=2)
#
#  Supported criteria are “gini” for the Gini impurity and “entropy” for the information gain.
#  n_estimators - trees number(iterations),  n_jobs - parallel execution of kernels in processor
#
rf.fit(X_train, y_train)
```

# Validation

```python
from sklearn.metrics import accuracy_score
y_pred = rf.predict(X_test)
print('accuracy - ', accuracy_score(y_test, y_pred))
```

# Predict

```python
prediction = rf.predict([[6, 3, 5, 2]])
print(target_names[prediction])
```

```python
print('For prediction enter features')
F = []
for name in feature_names:
  f = float(input(name))
  F.append(f)
prediction = rf.predict(np.asarray(F).reshape(1,-1))
print('Features: ', F, ' Predict - ', target_names[prediction])
```
