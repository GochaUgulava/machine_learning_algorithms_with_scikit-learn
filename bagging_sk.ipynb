```python
import numpy as np
import pandas as pd
```

# Data

```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
df_wine.columns = ['Label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
                        'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                         'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
df_wine.head()
```

```python
X = df_wine.iloc[:, 1:].values
y = df_wine.iloc[:, 0].values
print(np.shape(X), np.shape(y))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
print(np.shape(X_train), np.shape(X_test), np.shape(y_train), np.shape(y_test))
```

# Bagging with ensemble of 500 None max_depth Decision Trees

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='entropy', max_depth=None, random_state=1)
bag = BaggingClassifier(base_estimator=tree, n_estimators=500, 
                        max_samples=1.0, max_features=1.0, 
                        bootstrap=True, bootstrap_features=False,
                        n_jobs=1, random_state=1)
```

```python
from sklearn.metrics import accuracy_score
tree = tree.fit(X_train, y_train)
y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
tree_train_score = accuracy_score(y_train, y_train_pred)
tree_test_score = accuracy_score(y_test, y_test_pred)
print('Accuracy score for decision tree train/test %0.3f/%0.3f' % (tree_train_score, tree_test_score))
```

```python
bag = bag.fit(X_train, y_train)
y_train_pred = bag.predict(X_train)
y_test_pred = bag.predict(X_test)
bag_train_score = accuracy_score(y_train, y_train_pred)
bag_test_score = accuracy_score(y_test, y_test_pred)
print('Accuracy score for bagging train/test %0.3f/%0.3f' % (bag_train_score, bag_test_score))
```
