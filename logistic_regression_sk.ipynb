```python
import numpy as np
from sklearn import datasets
```

# Load Data

```python
iris = datasets.load_iris()
iris
```

```python
X = iris['data']
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
labels = np.unique(y)
labels
```

# Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
print(np.bincount(y))
print(np.bincount(y_train))
print(np.bincount(y_test))
```

# Normalize

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)            #  get mean and std from X_train, which used fo both - X_train and X-test
X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)
```

# Model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=100.0, random_state=1, solver='lbfgs', multi_class='multinomial')
#
#  solver - 'newton-cg', 'lbfgs', 'liblinear ', 'sag', 'saga'   
#  multi_class - 'ovr', 'multinomial', 'auto
#
lr.fit(X_train_std, y_train)
```

# Validation

```python
from sklearn.metrics import accuracy_score
y_pred = lr.predict(X_test_std)
print('accuracy - ', accuracy_score(y_test, y_pred))
```

# Predict

```python
lr.predict_proba(X_test_std[:3, :])
```

```python
lr.predict_proba(X_test_std[:3, :]).argmax(axis=1)
```

```python
prediction = lr.predict([[6, 3, 5, 2]])
print(target_names[prediction])
```

```python
print('For prediction enter features')
F = []
for name in feature_names:
  f = float(input(name))
  F.append(f)
prediction = lr.predict(scaler.transform(np.asarray(F).reshape(1,-1)))
print('Features: ', F, ' Predict - ', target_names[prediction])
```
